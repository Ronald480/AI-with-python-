```python
# Step 1: Importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.linear_model import LinearRegression


%matplotlib inline
```

```python
# Step 2: Performing regression with single input feature

number_of_samples = 100

# Generating the X values
X = np.linspace(-np.pi, np.pi, number_of_samples)

# Generating the corresponding Y values using the given function
def f(X):
    return X/2 + np.sin(X) + np.random.random(X.shape)

Y = f(X)

# Displaying the generated values of X and Y
print("X: ", X)
print("Y: ", Y)

# Plotting a scatter diagram of the X.Y variables
plt.scatter(X, Y, color='black')
plt.xlabel("x-input feature")
plt.ylabel("y-input feature")
plt.title("Fig 1: Data for linear regression for the diagram")
plt.show()
```

```python
# Step 3: Splitting the data into training, validation and testing sets
random_indices = np.random.permutation(number_of_samples)

# Training set(70% of the data)
x_train = X[random_indices[:70]]
y_train = Y[random_indices[:70]]

# Validation set(15% of the data)
x_val = X[random_indices[70:85]]
y_val = Y[random_indices[70:85]]

# Testing set(15% of the data)
x_test = X[random_indices[85:]]
y_test = Y[random_indices[85:]]


print(x_test.tolist())
print(y_test.tolist())
```

```python
# Step 4: Fitting a linear model
Model = LinearRegression() # Linear regression object (Model)

# Converting training model lists into column matrices
mod_x = np.asarray(x_train.reshape(len(x_train), 1))
mod_y = np.asarray(y_train.reshape(len(y_train), 1))

# Fiting a line to training model
Model.fit(mod_x, mod_y)

# Scatter diagram of X_train and Y_train (Basically the same as the original data)
plt.scatter(x_train, y_train)

# line = mx + c for the line graph using predict()
line = Model.predict(mod_x)

# Plot the line graph (Line of best fit / Regression line)
plt.title("Line graph on Scatter diagram")
plt.plot(x_train, line)
plt.show()


# Two column matrices (M, N)
M = mod_x   # X column of the dataset
N = Model.predict(M)    # Predicted Y column of the dataset


# Displaying the linear regression line
plt.title("Fitted Line")
plt.plot(M, N)
plt.show()
```

```python
# Step 5: Evaluating the model

# Mean error of the training set
y_val_pred = Model.predict(x_val.reshape(len(x_val), 1))    # Predicted values of the X_val of the validation set
mn_val_error = np.mean((y_val - y_val_pred)**2)    # Mean error of the validation set

# Mean error of the test set
y_test_pred = Model.predict(x_test.reshape(len(x_test), 1))  # Predicted values of the X_test of the test set
mn_test_error = np.mean((y_test - y_test_pred)**2 )    # Mean error of the test set

print("Mean error of the validation set: %.2f" % mn_val_error)
print("Mean error of the test set:  %.2f" % mn_test_error)
```
